### fetching the dataframe

```python
from sklearn.datasets import load_boston
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

boston_dataset = load_boston()
```

### creating data frame

```python
boston_df = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston_df
```

```python
boston_df['MEDV'] = boston_dataset.target
boston_df
```

```python
boston_df.isnull().sum()
```

```python
boston_df.shape
```

```python
X = boston_df.iloc[:,1:-1].values
y = boston_df.iloc[:,-1].values
```

```python
from sklearn import model_selection
Xtrain,Xtest,ytrain,ytest = model_selection.train_test_split(X,y, test_size =0.3, random_state =25)
```

### applying linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(Xtrain, ytrain)

ypred = lm.predict(Xtest)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
from math import sqrt
from sklearn.metrics import r2_score, mean_squared_error

print(sqrt(mean_squared_error(ytest,ypred)))
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
ridgereg = Ridge(alpha =0.001, normalize = True)
ridgereg.fit(Xtrain , ytrain)

ytrain_ridge = ridgereg.predict(Xtrain)
ytest_ridge =ridgereg.predict(Xtest)

print(sqrt(mean_squared_error(ytrain,ytrain_ridge)))
print(sqrt(mean_squared_error(ytest,ytest_ridge)))

print('R2 value/ Coefficient of determination: {}'.format(ridgereg.score(Xtest,ytest)))
```

### Lasso Regression

```python
from sklearn.linear_model import Lasso
lassoreg = Lasso(alpha =0.001, normalize = True)
lassoreg.fit(Xtrain , ytrain)

ytrain_lasso = lassoreg.predict(Xtrain)
ytest_lasso =lassoreg.predict(Xtest)

print(sqrt(mean_squared_error(ytrain,ytrain_lasso)))
print(sqrt(mean_squared_error(ytest,ytest_lasso)))

print('R2 value/ Coefficient of determination: {}'.format(lassoreg.score(Xtest,ytest)))
```

### Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
elasticreg = ElasticNet(alpha =0.001, normalize = True)
elasticreg.fit(Xtrain , ytrain)

ytrain_elastic = elasticreg.predict(Xtrain)
ytest_elastic =elasticreg.predict(Xtest)

print(sqrt(mean_squared_error(ytrain,ytrain_elastic)))
print(sqrt(mean_squared_error(ytest,ytest_elastic)))

print('R2 value/ Coefficient of determination: {}'.format(elasticreg.score(Xtest,ytest)))
```
